## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the variables
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Display the information
plt.plot(x, y, label="sin(x)")
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the variables
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Customize the sets
plt.plot(x, y1, label="sin(x)", color = "yellow")
plt.plot(x, y2, label="cos(x)", color = "blue")

# Display the sets
plt.title("Sine and Cosine Function")
plt.xlabel("x")
plt.ylabel("Funtion Value")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the variables
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

# Customize and display the sets
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

âœ… **Your Turn:** Generate two sets of random points and plot them in different colors on the same scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two sets
x1 = np.random.rand(50)
y1 = np.random.rand(50)

x2 = np.random.rand(50)
y2 = np.random.rand(50)

# Create the sizes
sizes = 300 * np.random.rand(50)

# Create scatter plot
plt.scatter(x1, y1, s=sizes, color="brown", label='Set 1')
plt.scatter(x2, y2, s=sizes, color="green", label='Set 2')

# Display the sets
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='lower right')
plt.show()
```

## 4. Bar Plot

```python
# Create the variables
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

# Display the sets
plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

## 5. Customizing Plots


```python
import numpy as np
import matplotlib.pyplot as plt

# Create the variables
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)


# Customize and display the sets
plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

âœ… **Your Turn:** Customize your own plot with labels, legend, and grid lines.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the line
x = np.linspace(0, 2*np.pi, 100)
y1 = np.tan(x)
y2 = np.cos(x)

# Customize the sets
plt.plot(x, y1, label="tan(x)", linestyle="--", color="pink")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="blue")

# Display the sets
plt.title("Customized Line Plot")
plt.xlabel("radians")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

### Summary
* `%matplotlib inline` makes plots display inside the notebook.
* Line, scatter, and bar plots are the most common basics.
* Titles, labels, legends, and grids improve readability.
* Visuals tell stories about data more effectively than raw numbers.
